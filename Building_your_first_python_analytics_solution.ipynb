```python
from IPython.core.magic import register_cell_magic
@register_cell_magic
def run_and_save(line, cell):
    'Run and save python code block to a file'
    with open(line, 'wt') as fd:
        fd.write(cell)
    code = compile(cell, line, 'exec')
    exec(code, globals())
```

```python
%%run_and_save myscript.py

def square(x):
    return x**2

for i in range(1,4):
    print('square of {} is {}'.format(i,square(i)))
```

```python
price_data = pd.read_csv('https://archive.org/download/weight_height/price.csv')
```

```python
price_data.head()
```

```python
price_data = price_data.fillna(0)
```

```python
price_data.head()
```

```python
def get_price_info():
    for index, row in price_data.iterrows():

        total_price = row['Total_Price']
        quantity = row['Quantity']

        try:
            price_of_a_unit = (total_price/quantity)
        except ZeroDivisionError:
            price_of_a_unit = 0
        print(price_of_a_unit)
```

```python
get_price_info()
```

```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
weight_height = 'https://archive.org/download/weight_height/weight_height.csv'
weight_height_df = h2o.import_file(path=weight_height)
```

```python
train, test = weight_height_df.split_frame(ratios=[0.85])
```

```python
x = train.columns
y = "Gender"
x.remove(y)
```

```python
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

```python
aml = H2OAutoML(max_models=100, seed=1, nfolds = 5)
aml.train(x=x, y=y, training_frame=train)
#aml.train(x=x, y=y, training_frame=train, validation_frame= valid)
```

```python
preds = aml.predict(test)
preds
```
